```python
import pandas as pd
import numpy as np

data = {
    'ID': [1, 2, 3, 4, 5],
    'NAME': ['Ram', 'Hari', 'Sita', 'Rushab', 'Geeta'],
    'AGE': [20, 25, 30, np.nan, 20],
    'ADDRESS': ['kathmandu', 'Pokhara', 'kathmandu', 'Pokhara', 'Butwal'],
    'SALARY': [50000, 60000, 30000, 10000, np.nan]
}

# Create DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
df
```

```python

print("Missing values before filling:\n", df.isnull().sum())
```

```python
df.head()
```

```python
df['AGE']=df['AGE'].fillna(df['AGE'].mode()[0])
df['SALARY']=df['SALARY'].fillna(df['SALARY'].mode()[0])
print("\nDataFrame after filling missing values with mode:\n")
df
```

```python
df['AGE'] = df['AGE'].fillna(df['AGE'].mean())
df['SALARY'] = df['SALARY'].fillna(df['SALARY'].mean())

print("\nDataFrame after filling missing values with mean:\n")
df
```

```python

df['AGE'] = df['AGE'].fillna(df['AGE'].median())
df['SALARY'] = df['SALARY'].fillna(df['SALARY'].median())

print("\nDataFrame after filling missing values with median:\n")
df
```

```python
df['BONUS'] = df['SALARY'] * 0.15
#Updated DataFrame
df
```

```python
grouped_df = df.groupby('ADDRESS')['SALARY'].mean().reset_index()
grouped_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x='ADDRESS', y='SALARY', data=grouped_df, palette='viridis')

plt.title('Average Salary by Address')
```

```python
plt.figure(figsize=(8, 6))  # Set figure size
sns.scatterplot(x='AGE', y='SALARY', data=df, hue='ADDRESS', palette='Set2', s=100)

# Customize the plot
plt.title('Scatterplot of Age vs Salary')  # Title of the plot
```
