```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ["Probability and Statistics", "Data Science and Machine Learning", "Artificial Intelligence", "Computer Graphics", "Engineering Economics", "Research Fundamentals"]

marks = [60, 45, 59, 72, 80, 91]
```

```python
plt.figure(figsize=(35, 15))
plt.subplot(2, 2, 1)
plt.plot(subjects, marks, marker='o', color='blue')
plt.title("Line Plot")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(16,8))
x = np.arange(len(subjects))
plt.scatter(x, marks, color='red')
plt.title("Scatter Plot")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(x, subjects)
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.bar(subjects, marks, color='green')
plt.title("Bar Graph")
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

```python
plt.pie(marks, labels=subjects, startangle=90, colors=['gold', 'lightblue', 'pink', 'lightgreen', 'purple', 'brown'])
plt.title("Pie Chart")
```
